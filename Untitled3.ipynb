```python
#Clustering - Univariate,Bivariate,Multivariate
```

```python
clustering1 = KMeans(n_clusters=6)
```

```python
clustering1.fit(df[['Annual_Income_(k$)']])
```

```python
clustering1.labels_
```

```python
df['Income Cluster'] = clustering1.labels_
df.head()
```

```python
df['Income Cluster'].value_counts()
```

```python
clustering1.inertia_
```

```python
intertia_scores=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df[['Annual_Income_(k$)']])
  intertia_scores.append(kmeans.inertia_)
```

```python
intertia_scores
```

```python
plt.plot(range(1,11),intertia_scores)
```

```python
df.groupby('Income Cluster')[['Age', 'Annual_Income_(k$)', 'Spending_Score']].mean()
```

```python
#Bivariate Clustering
```

```python
clustering2 = KMeans()
clustering2.fit(df[['Annual_Income_(k$)', 'Spending_Score']])
clustering2.labels_
df['Spending and Income Cluster'] = clustering2.labels_
df.head()
```

```python
inertia_scores2 = []
for i in range(1, 11):
    kmeans2 = KMeans(n_clusters=i)
    kmeans2.fit(df[['Annual_Income_(k$)', 'Spending_Score']])
    inertia_scores2.append(kmeans2.inertia_)
plt.plot(range(1,11),inertia_scores2)
```

```python
centers=pd.DataFrame(clustering2.cluster_centers_)
centers.columns = ['x','y']
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=centers['x'],y=centers['y'],s=100,c='black',marker='*')
sns.scatterplot(data=df, x='Annual_Income_(k$)', y='Spending_Score', hue='Spending and Income Cluster', palette='tab10')
```

```python
pd.crosstab(df['Spending and Income Cluster'],df['Genre'],normalize='index')
```

```python
df.groupby('Spending and Income Cluster')[['Age', 'Annual_Income_(k$)', 'Spending_Score']].mean()
```
